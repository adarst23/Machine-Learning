```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.datasets import make_circles
```

```python
X, y = make_circles(n_samples=200, noise=0.15, factor=0.1)
```

```python
plt.scatter(X[:,0],X[:,1], c=y, cmap='bwr')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear')

svc.fit(X_train,y_train)
pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(pred,y_test))
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X_test, y_test, svc, legend=2)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr')

# Get unique labels and colors from the scatter plot
unique_labels = np.unique(y)
scatter = plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr')
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=scatter.cmap(scatter.norm(label)), markersize=10, label=f'Label {label}') for label in unique_labels]

# Add a legend with the custom handles
plt.legend(handles=handles, title="Labels")
plt.show()
```

## Using the Radial Fuction Basis on the X

```python
z = np.exp(-(X**2)).sum(1)
ax = plt.subplot(projection='3d')
ax.scatter3D(X[:,0],X[:,1],z,c=y, cmap='bwr')
```

```python
rbf = SVC(kernel='rbf')

rbf.fit(X_train,y_train)

pred = rbf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test,pred))
```

```python
plot_decision_regions(X,y,rbf)
```
