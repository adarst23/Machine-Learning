```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8 ,random_state=101)
```

```python
X.shape
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='rainbow', edgecolor='black')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(X,y)
```

```python
kmeans.labels_
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, sharey=True, figsize= (10,7))

ax[0].set_title('Original')
ax[0].scatter(x=X[:,0], y=X[:,1], c=y, cmap='rainbow', edgecolor='black')

ax[1].set_title('KMeans')
ax[1].scatter(x=X[:,0], y=X[:,1], c=kmeans.labels_, cmap='rainbow', edgecolor='black')
```

It has clustered the groups almost correctly. Its a good clustering
